```python
import tensorflow as tf
import os

# Assure-toi que le dossier de destination existe
os.makedirs("model", exist_ok=True)

# Charger le modèle .keras original sans compile
model = tf.keras.models.load_model("model/best_fpn_mobilenet_categorical_crossentropy.keras", compile=False)

# Convertisseur TFLite
converter = tf.lite.TFLiteConverter.from_keras_model(model)

# Optionnel : optimisation (quantization, compression)
# converter.optimizations = [tf.lite.Optimize.DEFAULT]

# Conversion
tflite_model = converter.convert()

# Sauvegarde
with open("model/model.tflite", "wb") as f:
    f.write(tflite_model)

print("✅ Modèle converti en TFLite et sauvegardé dans model/model.tflite")
```
